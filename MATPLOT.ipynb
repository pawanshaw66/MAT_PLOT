```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x  = np.arange(0,10)
y  = np.arange(11,21)
```

```python
a  = np.arange(30,40)
b  = np.arange(40,50)
```

```python
# ploting using matplot
plt.scatter(x,y,c='g')         # g is color 
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'b--')
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'ro--',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'y')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'pink')
plt.subplot(2,2,4)
plt.plot(x,y,'g')
```

```python
x = np.arange(1,11)
y = 3*x + x*4
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x,y,'red')
```

```python
np.pi
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('Sine Wvae Form')
plt.plot(x,y,'orange')
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x,y_sin)
```

```python
plt.plot(x,y_cos)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
plt.title('cosine')
```

```python
x= [2,6,9]
y = [8,3,7]
plt.bar(x,y)

x1 = [3,25,14]
y1 = [2,11,22]
plt.bar(x1,y1,color='orange')

plt.title('Bar Graph')
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
a = np.array([1,5,10,5,21,25,3,9,7,4,15])
plt.hist(a,color='green')
plt.title('Histogram',color='green')
```

```python
# box plot
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# box plot
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=False,patch_artist=True);
```

```python
labels = 'Python','C++','Java','Javascript','Ruby'
sizes = [200,225,190,310,268]    # sizes in cumulative value   ex : 200/(200+225+190+310+268)=_%
colors = ['lightgreen','red','gold','pink','lightblue']
explode =[0.1,0,0,0,0]
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)    # autopct show in which formate we show the value
plt.axis('equal')

```
